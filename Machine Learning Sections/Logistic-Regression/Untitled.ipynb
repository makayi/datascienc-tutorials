```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if(pd.isnull(Age)):

        if Pclass==1:
            return 37
        elif Pclass==2:
            return 27
        elif Pclass==3:
            return 25
    else:
        return Age
```

```python
titantic_data['Age']=titantic_data[['Age','Pclass']].apply(impute_age,axis=1)      
```

```python
sns.heatmap(titantic_data.isnull(),yticklabels=False,cmap='viridis')
```

```python
from sklearn.model_selection  import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
titantic_data.drop(['Embarked'],axis=1,inplace=True)
```

```python
X=titantic_data[['Pclass','male','SibSp','Age','Fare','Parch']]
y=titantic_data['Survived']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=101)
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
def map_to_fate(predictions):
    preds=[]
    for pred in predictions:
        if(pred==0):
            preds.append({'fate':'didnt survive'})
        if(pred==1):
            preds.append({'fate':'survived'})
    return preds
            
```

```python
preds=map_to_fate(predictions)
preds
```
